### Analysis and Questions

#### Most Popular, Admired, and Desired Technologies

We lost about 20,000 respondants from that. Sorry students and hobbyists ;-; . Now, let's move on to answering some interesting questions. Looking at the columns, there are a few columns that give insight into which platforms/languages/technologies are most admired, most worked with, and most sought after. So, we want to find out in each category what language/platform/technology was *MOST* sought after, most used, and most admired in 2024. 

```python
df
```

```python
for subject in ["Platform", "Language", "Webframe"]:
    admired_col = subject + "Admired"
    worked_with_col = subject + "HaveWorkedWith"
    desired_col = subject + "WantToWorkWith"
    cols_to_keep = [admired_col, worked_with_col, desired_col]
    curr_df = df[cols_to_keep]

    admired_count_groups = curr_df.groupby(admired_col).count()
    admired_count_groups = admired_count_groups.drop(admired_count_groups.columns[0], axis=1)
    most_admired = admired_count_groups.idxmax().iloc[0]

    worked_with_count_groups = curr_df.groupby(worked_with_col).count()
    worked_with_count_groups = worked_with_count_groups.drop(worked_with_count_groups.columns[0], axis=1)
    most_worked_with = worked_with_count_groups.idxmax().iloc[0]

    desired_count_groups = curr_df.groupby(desired_col).count()
    desired_count_groups = desired_count_groups.drop(desired_count_groups.columns[0], axis=1)
    most_desired = desired_count_groups.idxmax().iloc[0]

    print(f"Most desired {subject}: {most_desired}")
    print(f"Most admired {subject}: {most_admired}")
    print(f"Most worked with {subject}: {most_worked_with}")
```

Boom! Nine questions answered with one loop. Pandas is pretty cool! So, we found out that AWS is absolutely dominating the developer platform space, which is consistent with what we are seeing on developer forms and professional networking sites like LinkedIn. It seemed that everyone was using AWS, and now the data just confirmed that trend of AWS being dominant.

In the language space, it seems that everyone wants to work with Python, and everyone admires Python. This is kind of surprising to us, given that Rust has been the talk of the town with it's speed and memory safety. I'd expect given those facts about it, and it's steep learning curve, that it'd be most admired - but given that Python is the language of data, it's makes sense that it's on top. Then of course everyone knows HTML, CSS, and JS/TS because most devs work with web technologies. Then web frameworks, no doubt React has been thee framework for a while. No surprises there. Though, the dominance of Spring Boot in popularity is a bit surprising to us.

#### The Economics of Software Developers

It's no secret that software development is quite a lucrative career. Did it stay that way in 2024, after the explosion of LLMs? Probably. But let's check with the "total compensation" data. For this section of the analysis, we are going to focus on developers that live in the US, given that we live in the US and are most familiar with the US economy. 

```python
econ_df = df[df['Country'].str.contains("United States")]
econ_df.shape
```

We still have almost 9000 developers to analyze even when limiting our scope to the US, which is nice. Let's continue by making sure that the currency being used is USD, since for whatever reason a developer might have entered their compensation in another currency.

```python
econ_df[econ_df['Currency'].str.contains("United States")]
```

Now, let's take a look as a baseline for salary what the average salary is like for a developer. Were developers still compensated above the average person in the US? (around 60k USD) ** *note that salary and total comp are different, but we are simplifying here*

```python
econ_df['CompTotal'].mean()
```

Well, we knew developers were well compensated, but $1,214,763,076 total annual compensation is a bit suspicious. There are probably some joke responses in here, so let's filter outliers by using some basic statistics about the compensation information.

```python
comp_df = econ_df['CompTotal']
comp_df.describe()
```

It looks like our 25th%ile is 108k USD, our 75th%ile is 200k USD, our max 7.7e12 or 7 trillion USD, and our min 0 USD. Given that information, we can make an informed decision to use 1 million USD as the max a developer might make, and around 20k as our minimum.

```python
comp_df = comp_df[(comp_df > 20000) & (comp_df < 1000000)]
comp_df.mean()
```

There we go, that makes much more sense. Now with our adjusted compensation data, we can see that we have an average annual total compensation of 167k USD for a developer in 2024. This makes it clear that developers still are very well compensated.

Let's visualize the salary information with a histogram.

```python
import matplotlib.pyplot as plt

comp_df.hist(bins=30)
plt.title("US Developer Compensation Histogram ($20k < comp < $1m)")
plt.xlabel("Total Annual Compensation (USD)")
plt.ylabel("# of developers")
plt.show()
```

The histogram provides a nice visual that shows that the majority of US developers make less than $200k total annual compensation, which explains our mean. Despite our mean, there are some outliers making a whole lot more than 160k total comp, with some *reportedly* making more than $900k USD. While not impossible in spaces like quantitative trading, it's reasonable to conclude that making that much money as a developer annual is extremely rare.

Now then, what about education and total compensation, surely there must be a link there? Let's see what the data says.

```python
econ_df = econ_df[(econ_df['CompTotal'] > 20000) & (econ_df['CompTotal'] < 1000000)]
econ_df.groupby("EdLevel")['CompTotal'].mean()
```

Unsurprisingly, developers with an undergraduate or graduate degree (Bachelor's / Master's / Doctorate or Professional) had higher average total compensations. On the other hand, we were surprised to see that developers with an Associate degree were compensated less on average than developers who only completed secondary school or those that dropped out of college/university. In either case, it's clear that despite not being highly educated in an academic way, the average developer is making great money.
