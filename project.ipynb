### Wrangling The Data
Now that we have our data loaded, we have to wrangle it. There is a lot of information here, we don't need all of it. Let's first start with seeing what the column situation looks like.

```python
for c in df.columns: print(c) #using this so we can expand and view as scrollable element
print(df.shape)
```

114 columns is too much data to be working with for our intents and purposes, so we are going to drop some of the columns that won't aid in our analysis. We are going to make use of the `survey_results_schema.csv` file that is part of the dataset on Kaggle that describes what the more obscure columns are.

Looking at the schema, we quickly can see that the "knowledge" columns won't be of use for us in this analysis. These columns are talking about work-place specific resources and procedures. We can drop these.

There are many more columns in here that aren't going to be useful, so we are going to loop through each column and drop them accordingly.

```python
for c in df.columns:
    if "Knowledge" in c or "JobSat" in c or "Frequency" in c or "OfficeStack" in c or "Professional" in c:
        df = df.drop(c, axis=1)

print("Num columns after drop:", df.shape[1])
for c in df.columns: print(c)
```

We have dropped around 30 columns, and the columns that currently exist give rise to some interesting questions! But before we get started with that, we are going to limit our analysis to *professional developers only* to gain insight into what the landscape of the professional software development industry looks like.

```python
df = df[df['Employment'].str.contains("Employed")]
df.shape
```

We lost about 20,000 respondants from that. Sorry students and hobbyists ;-; . Now, let's move on to answering some interesting questions. Looking at the columns, there are a few columns that give insight into which platforms/languages/technologies are most admired, most worked with, and most sought after. So, we want to find out in each category what language/platform/technology was *MOST* sought after, most used, and most admired in 2024. 

```python
df
```

```python
for subject in ["Platform", "Language", "Webframe"]:
    admired_col = subject + "Admired"
    worked_with_col = subject + "HaveWorkedWith"
    desired_col = subject + "WantToWorkWith"
    cols_to_keep = [admired_col, worked_with_col, desired_col]
    curr_df = df[cols_to_keep]

    admired_count_groups = curr_df.groupby(admired_col).count()
    admired_count_groups = admired_count_groups.drop(admired_count_groups.columns[0], axis=1)
    most_admired = admired_count_groups.idxmax().iloc[0]

    worked_with_count_groups = curr_df.groupby(worked_with_col).count()
    worked_with_count_groups = worked_with_count_groups.drop(worked_with_count_groups.columns[0], axis=1)
    most_worked_with = worked_with_count_groups.idxmax().iloc[0]

    desired_count_groups = curr_df.groupby(desired_col).count()
    desired_count_groups = desired_count_groups.drop(desired_count_groups.columns[0], axis=1)
    most_desired = desired_count_groups.idxmax().iloc[0]

    print(f"Most desired {subject}: {most_desired}")
    print(f"Most admired {subject}: {most_admired}")
    print(f"Most worked with {subject}: {most_worked_with}")
```
